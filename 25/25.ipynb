```python
with open("input.txt", "r") as f:
    lines = f.read().split("\n\n")

keys = [line for line in lines if line[:5] == "....."]
locks = [line for line in lines if line[:5] == "#####"]

lock_signatures = set()

for lock in locks:
    signature = []
    m = lock.splitlines()
    for x in range(len(m[0])):
        counter = 0
        for y in range(len(m)):
            if m[y][x] != "#":
                break
            counter += 1
        signature.append(counter - 1)

    lock_signatures.add(tuple(signature))


existing_keys = 0

for key in keys:
    signature = []
    m = key.splitlines()

    for x in range(len(m[0])):
        counter = 0
        for y in range(len(m) - 1, -1, -1):
            if m[y][x] != "#":
                break
            counter += 1
        signature.append(counter - 1)

    for lock_signature in lock_signatures:
        for l, k in zip(lock_signature, signature):
            if k > 5-l:
                break
        else:
            existing_keys += 1

print(existing_keys)
```
